#Due to the large amount of data ,I`m going to preprocess each node seperately to make sure that preprocessing done well 

#Node 1

```python
df1= pd.read_csv('csvData_1.csv')
df1
```

```python
# Get info about dataset
df1.info()
```

```python
#convert "snapshot_datetime" datatype to be a datatime datatype 
df1['snapshot_datetime']=pd.to_datetime(df1['snapshot_datetime'])
```

```python
# Drop unwanted columns
df1.drop(['meta-data','destination_request_id'], axis=1, inplace=True)
df1
```

```python
#Filling NULL values
df1[['collector_id']] = df1[['collector_id']].fillna(value=51)
df1[['device_id']] = df1[['device_id']].fillna(method="ffill")
df1[['latitude']] = df1[['latitude']].interpolate()
df1[['longitude']] = df1[['longitude']].interpolate()
df1[['snapshot_datetime']]=df1[['snapshot_datetime']].fillna(method="ffill")
df1
```

```python
#Reorder columns 
df1 = df1[['collector_id', 'device_id', 'latitude', 'longitude','snapshot_datetime']]
df1
```

```python
# Get info again after preprocessing 
df1.info()
```

```python
#Making sure that there isn1t any null values 
df1.isna().sum()
```

```python
#drop duplicate data instances if any
df1.drop_duplicates(inplace=True)
df1.shape
```

# Now , using GeoPandas dataframe to deal with map locations efficiently. 

```python
# Select the fourth node to deal with its locations because it is just 84 record so it is possiable to visualize it 
gdf = gpd.GeoDataFrame(df4, geometry=gpd.points_from_xy(df4.longitude, df4.latitude))
gdf
```

# The next two cells visualize the locations on simple graph 

```python
# On small scale 
f, ax = plt.subplots(figsize=(6, 6))
gdf.plot(ax=ax)
ax.set_axis_off()
plt.show()
```

```python
# On large scale for more Clearness 
gdf.plot(figsize=(20, 20))
plt.show()
```

```python
import folium
#select random location from node 4 (fourth dataframe)
map = folium.Map(location=[30.562304 ,31.560093])
map
```

#Add points (locations) to map

```python
for _, r in gdf.iterrows():
    # Without simplifying the representation of each borough, the map might not be displayed.
    sim_geo = gpd.GeoSeries(gdf['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j, style_function=lambda x: {'fillColor': 'orange'})
    folium.Popup(gdf['snapshot_datetime']).add_to(geo_j)
    geo_j.add_to(map)
map
```
