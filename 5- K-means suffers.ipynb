```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial import distance
from scipy import linalg
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.cluster import SpectralClustering
from sklearn.neighbors import kneighbors_graph
from sklearn.metrics import silhouette_score
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.metrics.cluster import fowlkes_mallows_score
import warnings
warnings.filterwarnings('ignore')
```

```python
moons=datasets.make_moons(n_samples=1000, noise=0.05,random_state=42)
circles=datasets.make_circles(n_samples=1000,noise=.02,factor=0.5,random_state=42)
datasets=[moons,circles]
```

```python
def plot_data(data,y,title):
  sns.scatterplot(data[0][:,0],data[0][:,1],hue=y)
  plt.title(title)
  plt.show()
```

```python
plot_data(moons,moons[1],'Orignal moons dataset')
plot_data(circles,circles[1],'Orignal circles dataset')
```

```python
Kmean=KMeans(2,init='k-means++')
Spectral=SpectralClustering(2,affinity='nearest_neighbors',assign_labels='kmeans',random_state=42)
DBscan=DBSCAN(eps=0.3,min_samples= 2)
Agglomerative = AgglomerativeClustering(2,linkage='single')
cluster_methods=[Kmean,Spectral,DBscan,Agglomerative]
methods=['Kmean','Spectral','DBscan','Agglomerative']
Dsets=['Moons','Circles']

#algos=[KMeans(2,init='k-means++'),SpectralClustering(2,affinity='nearest_neighbors',assign_labels='kmeans',random_state=42),DBSCAN(eps=0.3,min_samples= 2)]
```

```python
def fu_(datasets,cluster_methods):
  j=0
  for dataset in datasets :
    i=0
    for algo in cluster_methods :
      print('-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-')
      print(methods[i])
      y_pred=algo.fit_predict(dataset[0]).copy()
      print('fowlkes_mallows_score= ',fowlkes_mallows_score(dataset[1], y_pred))
      print('adjusted_rand_score= ',adjusted_rand_score(dataset[1], y_pred))
      print('-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-')
      cap='data set name : '+ Dsets[j]+' after apply : '+ methods[i] +' clustering'
      plot_data(dataset,y_pred,cap)
      i+=1
    j+=1
```

```python
fu_(datasets,cluster_methods)
```
